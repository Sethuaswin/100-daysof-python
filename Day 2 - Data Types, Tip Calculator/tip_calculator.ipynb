### 3. Mathematical Operations in Python

* `+` - Addition
* `-` - Subtraction
* `*` - Multiplication
* `/` - Divison (it will return float data type always)
* `**` - Exponents (2^2)

````python
# Division will return float
print(6/3)
````

````python
# checking the data type of the division
print(type(6/3))
````

````python
# exponents
print(2**3)
````

The Mathematical opertaion will follow the `PEMDASLR` Rule

* Parentheses - `()`
* Exponents - `**`
* Multiplication - `*`
* Division - `/`
* Addition - `+`
* Subtraction - `-`
* Left - the multiplication and division are same priority which ever is in the left that will comuted first the same will be applicable to addition to addition and subtraction
* Right

````python
# what will be output of the below code

print(3 * 3 + 3 / 3 - 3)
````

The Order will be

* Multiplication - `3 * 3`
* Division - `3 / 3`
* Addition - `(3 * 3 = 9) + (3 / 3 = 1.0) = 10.0`
* Subtraction - `10.0 - 3 = 7.0`

````python
# how to bring the out put as 3.0 instead of 7.0
print(3 * (3 + 3) / 3 - 3)
````

````python
# if the bill was $150.00, split between 5 people, with 12% tip
# Each person should pay (150.00 / 5) * 1.12
# Round the each person has to pay with two decimal numbers

print("Welcome to the tip calculator")

total_bill = input("What was the total bill? $")
perc = input("What percentage tip would you like to give? 10, 12, or 15? ")
no_people = input("How many people to split the bill")

# converting the data types appropriatly
total_bill = float(total_bill)

# percentage addtion to 1.xx
perc = float("1."+perc)
no_people = int(no_people)

final_pay = (total_bill / no_people) * perc

# Rounding with two decimals
final_pay = round(final_pay,2)

print(f"Each person should pay: ${final_pay}")
````
