##### Question 1:

Which version of code will produce an Indentation Error when it is run?

NOTE: The • symbol represents a single space.

* ```python
def my_function():
••print("Hello")
```
* ```python
def my_function():
print("Hello")
```
* ```python
def my_function():
••••print("Hello")
```
* ```python
def my_function():
••print("Hello")
print("Bye")
```

**Answer**<br>
```python
def my_function():
print("Hello")
```
