### How to modify a Global variable



````python
################### Scope ####################

enemies = 1 # Global Scope

def increase_enemies():
    global enemies # calling the global variable
    enemies += 1  # Loacal Scope
    print(f"enemies inside function: {enemies}")

increase_enemies()
print(f"enemies outside function: {enemies}")

# These two variables are entirely two different variables
# Best practice is don't modify the global variable inside a function
````

````python
################### Scope ####################

enemies = 1 # Global Scope

def increase_enemies():
    print(f"enemies inside function: {enemies}")
    return enemies + 1

enemies = increase_enemies() # By this way we can modify the global variable with out modifying them
print(f"enemies outside function: {enemies}")
````

````python
# Own version
import random
print(logo)
print("Welcome to the NumberGuessing Game! ")
print("I'm Thinking of a number between 1 and 100")
answer = random.randint(1,100)
print(f"Pssst. the answer is {answer}")
difficulty = input("Choose a difficulty. Type 'easy' or 'hard': ")

if difficulty.lower() == "hard":
    turns = 5
elif difficulty.lower() == "easy":
    turns = 10
else:
    print("You have entered an invalid input!")

while turns > 0:
    print(f"You have {turns} attempts remaining to guess the number.")
    guess = int(input("Make a guess: "))
    if guess > answer:
        print("Too High")
        print("Guess again.")
    elif guess < answer:
        print("Too low")
        print("Guess again.")
    else:
        print(f"You got it! The answer was {answer}")
        turns = 0
    turns -= 1

if turns == 0:
    print("You have run out of guesses")
````

````python
# Lecturer Version

from random import randint

EASY_LEVEL_TURNS = 10
HARD_LEVEL_TURNS = 5

# Functon to set difficulty
def set_difficulty():
    level = input("Choose a difficulty. Type 'easy' or 'hard': ")
    if level.lower() == "hard":
        return HARD_LEVEL_TURNS
    else:
        return EASY_LEVEL_TURNS

# Function to check user's guess against actual answer
def check_answer(guess,answer, turns):
    """
    Check answer against guess. Returns the number of turns remaining.
    """
    if guess > answer:
        print("Too high.")
        return turns - 1
    elif guess < answer:
        print("Too low.")
        return turns - 1
    else:
        print(f"You got it! The answer was {answer}")


def game():
    print(logo)
    # Choosing a random number between 1 and 100
    print("Welcome to the NumberGuessing Game! ")
    print("I'm Thinking of a number between 1 and 100")
    answer = randint(1,100)
    print(f"Psst, the correct answer is {answer}")

    #Let the user guess a number
    turns = set_difficulty()

    # Repeat the guessing functionality if they get it wron
    guess = 0

    while guess != answer:
        print(f"You have {turns} attempts remaining to guess the number")
        guess = int(input("Make a guess: "))
        turns = check_answer(guess,answer,turns)
        if turns == 0:
            print("You have run out of guesses")
            return
        elif guess != answer:
            print("Guess Again")


# Track the number of turns and reduce by 1 if the get it wrong
game()
````
