## <center> Rock Paper Scissoors Challange </center>

````python
rock = '''
Rock
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
Paper
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
Scissors
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

#Write your code below this line 👇

game = [rock,paper,scissors]

player_choice = int(input('What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors\n'))
if player_choice >=3 or player_choice <0:
  print("You have typed invalid Number! You Loose!")
else:
  print(f"You Chose: \n{game[player_choice]}")

  import random
  computor_choice = random.randint(0,len(game)-1)

  print(f"Computor Chose: \n{game[computor_choice]}")

  if player_choice == computor_choice:
    print("The Game is Draw! Try Again")
  elif ((player_choice == 0) & (computor_choice == 1)) or ((player_choice == 1) & (computor_choice == 2)) or ((player_choice == 2) & (computor_choice == 0)):
    print("You Loose")
  else:
    print("You Won the Game! Congrats")
````
