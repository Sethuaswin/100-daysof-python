* However when the number is divisible by 3 then instead of printing the number it should print "Fizz".

* When the number is divisible by 5, then instead of printing the number it should print "Buzz".

```python
target = int(input())
for number in range(1, target + 1):
    ########### BUG ##################
    # if number % 3 == 0 or number % 5 == 0:
    ###################################
    if number % 3 == 0 and number % 5 == 0:
        print("FizzBuzz")
    elif number % 3 == 0: # Bug if is there instead of elif
        print("Fizz")
    elif number % 5 == 0: # Bug if is there instead of elif
        print("Buzz")
    else:
        print(number) # [] should be removed
```
